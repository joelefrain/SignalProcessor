# 05 - Ajuste espectral rapido

Aplica un ajuste aproximado en frecuencia para acercar el espectro del registro al objetivo.

```python
from pathlib import Path
import sys

ROOT = Path.cwd()
if not (ROOT / "src").exists():
    ROOT = ROOT.parent
SRC = ROOT / "src"
if str(SRC) not in sys.path:
    sys.path.insert(0, str(SRC))

import signalprocessor
print(f"ROOT={ROOT}")
print(f"signalprocessor={Path(signalprocessor.__file__).resolve()}")
```

```python
from signalprocessor.io import read_motion_csv, read_spectrum_csv, write_json, write_motion_csv, write_spectrum_csv
from signalprocessor.metrics import motion_metrics
from signalprocessor.plotting import save_spectrum_plot
from signalprocessor.scaling import frequency_domain_spectral_match

OUT = ROOT / "examples" / "ouputs" / "case_05_ajuste_espectral_rapido"
OUT.mkdir(parents=True, exist_ok=True)

motion = read_motion_csv(ROOT / "examples" / "data" / "motion" / "LIMAEW.csv", unit="g", name="LIMAEW")
target_t, target_sa = read_spectrum_csv(ROOT / "examples" / "data" / "response_spectrum" / "EPU_475.csv")
ouput = frequency_domain_spectral_match(motion, target_t, target_sa, iterations=3, max_factor_per_iteration=1.6, smoothing_width=7, highpass_hz=0.08)

write_motion_csv(OUT / "LIMAEW_ajuste_frecuencia.csv", ouput.motion, unit="g")
write_json(OUT / "LIMAEW_ajuste_frecuencia.json", {"method": ouput.method, "metrics": motion_metrics(ouput.motion)})
write_spectrum_csv(OUT / "LIMAEW_ajuste_frecuencia_espectro.csv", ouput.periods, ouput.scaled_sa_g)
save_spectrum_plot(OUT / "LIMAEW_ajuste_frecuencia.png", ouput.periods, {"ajustado": ouput.scaled_sa_g, "objetivo": ouput.target_sa_g}, title="Ajuste espectral rapido en frecuencia")

{"method": ouput.method, "output": str(OUT)}
```
