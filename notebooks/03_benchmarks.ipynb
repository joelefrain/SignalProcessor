# 03 - Benchmarks SignalProcessor

Compara SignalProcessor contra los archivos de referencia disponibles en `examples/data/benchmark`.

```python
from pathlib import Path
import sys

ROOT = Path.cwd().parent if Path.cwd().name == 'notebooks' else Path.cwd()
sys.path.insert(0, str(ROOT / 'src'))

import pandas as pd
import signalprocessor as sp
```

## Benchmark de correccion

Usa registros SMC no corregidos y los compara contra las versiones corregidas de referencia.

```python
correction_bench = sp.benchmark_correction(ROOT / 'examples')
correction_bench
```

## Benchmark de escalamiento

Compara metodos internos y, si existe, reporta el error espectral de archivos escalados de referencia.

```python
scaling_bench = sp.benchmark_scaling(
    ROOT / 'examples',
    records=('LIMANS', 'LIMAEW'),
    methods=('linear', 'frequency', 'wavelet'),
    period_range_s=(0.05, 2.0),
)
scaling_bench
```

## Exportar tablas

```python
out = ROOT / 'examples/output/benchmark'
out.mkdir(parents=True, exist_ok=True)
correction_bench.to_csv(out / 'correction_benchmark.csv', index=False)
scaling_bench.to_csv(out / 'scaling_benchmark.csv', index=False)
out
```
