# 06 - Benchmark contra SeismoMatch

Compara el escalamiento/ajuste espectral rapido de `SignalProcessor`, sin correccion de linea base ni filtrado, contra los acelerogramas exportados por SeismoMatch en `examples/data/benchmark`.

Criterio usado: error relativo medio entre espectros de respuesta en el rango `0.1-2.0 s`, con tolerancia `0.3`.

```python
from pathlib import Path
import sys

ROOT = Path.cwd()
if not (ROOT / "src").exists():
    ROOT = ROOT.parent
SRC = ROOT / "src"
if str(SRC) not in sys.path:
    sys.path.insert(0, str(SRC))

import signalprocessor
print(f"ROOT={ROOT}")
print(f"signalprocessor={Path(signalprocessor.__file__).resolve()}")
```

```python
from signalprocessor.benchmark import (
    plot_benchmark_spectra,
    run_seismomatch_benchmark,
    write_benchmark_report,
    write_scaled_benchmark_motions,
)

OUT = ROOT / "examples" / "output" / "case_06_benchmark_seismomatch"
run = run_seismomatch_benchmark(
    ROOT,
    tolerance=0.03,
    period_range=(0.1, 2.0),
    comparison_metric="rms_relative_error",
    method="log_least_squares",
    damping=0.05,
    beta=0.25,
    gamma=0.5,
    num_periods=200,
    iterations=50,
)

write_benchmark_report(run, OUT)
write_scaled_benchmark_motions(run, OUT / "motions", root=ROOT)
plot_benchmark_spectra(run, OUT / "plots")

summary = run.rows()
print(f"Todos aprobados: {run.all_passed}")
summary
```

La comparacion principal se hace contra el espectro de los acelerogramas de SeismoMatch, no contra muestra a muestra. Dos algoritmos de ajuste espectral pueden producir series temporales distintas aunque sus espectros sean compatibles.
